# Transformada de Domínio do Tempo para Frequência
Para transformada do domínio do tempo para frequência, utiliza-se a função da transformada rápida de Fourier.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

# Exemplo 01:

$\textbf{Dados:}$

Fa = 1000  - Frequência de amostragem  
T = 1.0/Fa - Período de amostragem  
f1 = 50.0  - Frequência do sinal 1 em Hz  
f2 = 100.0 - Frequência do sinal 2 em Hz  
a1 = 1.0   - amplitude do sinal 1  
a2 = 0.5   - amplitude do sinal 2  

```python
Fa = 1000  # Frequência de amostragem
T = 1.0/Fa # Período de amostragem
f1 = 50.0 #Frequência do sinal 1 em Hz
f2 = 100.0 #Frequência do sinal 2 em Hz
a1 = 1.0 #amplitude do sinal 1
a2 = 0.5 #amplitude do sinal 2
x = np.linspace(0.0, N*T, Fa)
y = a1*np.sin(f1 * 2.0*np.pi*x) + a2*np.sin(f2*2.0*np.pi*x)

plt.plot(x,y)
plt.show()
plt.grid()
plt.xlabel("Frequência")
plt.ylabel("Amplitude")

plt.title("Senóide no domínio da frequência")

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), Fa//2)
plt.plot(xf, 1.0/Fa * np.abs(yf[0:Fa//2]))
plt.show()
plt.grid()
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title("Senóide no domínio do tempo")
```

# Exemplo 02:

```python
Fa = 400  # Frequência de amostragem
T = 1.0/Fa # Período de amostragem
f1 = 10.0 #Frequência do sinal 1 em Hz
f2 = 20.0 #Frequência do sinal 2 em Hz
a1 = 2.0 #amplitude do sinal 1
a2 = 4.0 #amplitude do sinal 2
x = np.linspace(0.0, Fa*T, Fa)
y = a1*np.sin(f1 * 2.0*np.pi*x) + a2*np.sin(f2*2.0*np.pi*x)

plt.plot(x,y)
plt.show()
plt.grid()
plt.xlabel("Frequência")
plt.ylabel("Amplitude")

plt.title("Senóide no domínio da frequência")

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), Fa//2)
plt.plot(xf, 1.0/Fa * np.abs(yf[0:Fa//2]))
plt.show()
plt.grid()
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title("Senóide no domínio do tempo")
```

# Aliasing

$\textbf{Dados:}$

Fa = 60   - Frequência de amostragem  
Fa2 = 26    - Frequência limite
T = 1.0/Fa - Período de amostragem  
f1 = 2.0   - Frequência do sinal   
a1 = 1.0   - amplitude do sinal 1  

```python
Fa = 60 # Frequência de amostragem
Fa2 = 26 #Frequência com aliasing
T = 1.0/Fa # Período de amostragem
a1 = 1 #amplitude do sinal 1
f1 = 2
N = 40 #Número de amostras
x = np.linspace(0.0, Fa*T, Fa)
y = a1*np.sin(2*np.pi*f1*x)

plt.plot(x,y)

m = np.linspace(0.0, Fa*T,Fa2)
n = a1*np.sin(2*np.pi*f1*m)

plt.plot(m,n)

plt.show()
plt.grid()
plt.xlabel("Frequência")
plt.ylabel("Amplitude")
plt.title("Senóide no domínio da frequência")
plt.rcParams["figure.figsize"] = [12,4]

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), Fa//2)
plt.plot(xf, 1.0/Fa * np.abs(yf[0:Fa//2]),linestyle='--', marker='s',)
plt.show()
plt.grid()
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title("Senóide no domínio do tempo")

nf = 1.0/Fa * np.abs(yf[0:Fa//2])
xn = xf
yn = np.zeros(30)

plt.plot(xn,yn, color= '#1E90FF')
plt.vlines(2.06,0,0.4983, color='#1E90FF')
plt.show()
```
